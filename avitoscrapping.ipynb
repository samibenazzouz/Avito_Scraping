```python
# -*- coding: utf-8 -*-
"""
Created on Sat Nov 23 16:36:02 2019

@author: Sami
"""

from requests import get
from bs4 import BeautifulSoup
import requests
import numpy as np
url="https://www.avito.ma/fr/maroc/voitures-%C3%A0_vendre"
#print("L'url etudié est :",url)


    
def get_urls(n):
    urls=[]

    for i in range(1,n+1):
        
        var_url=['https://www.avito.ma/fr/maroc/voitures-%C3%A0_vendre?o={}'.format(i)]
        urls=urls+var_url
    return urls
        
URLS=get_urls(10)
print(URLS)
num_item=0
VOITURES=[]
for my_url in URLS:
    print("============================================")
    resp=requests.get(my_url)
    html_soup=BeautifulSoup(resp.text,'html.parser')
    vue=html_soup.find('div',class_="listing listing-thumbs")
    
    for i in range(num_item+1,num_item+36):
        li_item="li-item-{} ".format(i)
        vue1=vue.find('div',id=li_item)
        vue2=vue1.find('div',class_="ctext3 fs12")
        vue3=vue2.find('h2',class_="fs14")
        vue4=vue3.find('a',href=True)
        voiture_url=vue4.get('href')
        titre_voiture=vue4.text
        #print(titre_voiture)
        #print(voiture_url)
        
        # Nous sommes dans la page de chaque voiture
        
        
        try:
            req_voit=requests.get(voiture_url)
            print("OKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKK1/3")
            
            #Extraction du nombre de vues
            try:
                html_soup1=BeautifulSoup(req_voit.text,'html.parser')
                vue1=html_soup1.find('div',class_="panel-heading panel-heading-small").text
                str1=vue1
                v=[int(s) for s in str1.split() if s.isdigit()]
                vues_voit=v[0]
                #print(vues_voit)
                print("OKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKK2/3")
            except:
                vues_voit=np.nan
                print("nop")
                
            #Extraction du prix de la voiture
            try:
                prix=html_soup1.find('span',class_="amount value").text
                prix=prix.replace(" ","")
                str1=prix
                p=[int(s) for s in str1.split() if s.isdigit()]
                print("OKKKKKKKKKKKKKKKKKKKKKKKKKKKKKKK3/3")
            except:
                p=[np.nan]
                print("nopp")
                
            try:
                html_sam=BeautifulSoup(req_voit.text,'html.parser')
                vuesam=html_sam.find('div',class_="panel-body list-big list-big-icons panel-body-smallpadding")
                vuesam=vuesam.find('h2',class_="font-normal fs13 lh30 no-margin").text
                ville_voit=vuesam
            except:
                ville_voit=np.nan
            #extraction de l'annee de mise en circulation
            try:
                html_sam=BeautifulSoup(req_voit.text,'html.parser')
                vuesam=html_sam.find('aside',class_="panel panel-body panel-info")
                INFOS=[]
                for tag in vuesam.find_all('li'):
                    print (tag.text)
                    INFOS+=[tag.text.replace("\n","")]
                del(INFOS[7])
                del(INFOS[6])
                for i in range(0,len(INFOS)):
                               INFOS[i]=INFOS[i].split(":")[1]
                #0 annee modele voiture
                #1 kilometrage
                #2 essence ou diesel
                #3 marque
                #4modele
                #5puissance fiscale
                annee_voiture=INFOS[0]
                kill_voiture=INFOS[1]
                carb_voiture=INFOS[2]
                marque_voiture=INFOS[3]
                modele_voiture=INFOS[4]
                puissfisc_voiture=INFOS[5]
                print(INFOS)
            except:
                annee_voiture=np.nan
                kill_voiture=np.nan
                carb_voiture=np.nan
                marque_voiture=np.nan
                modele_voiture=np.nan
                puissfisc_voiture=np.nan
            
            
            
            
            
        except:
            print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
        
            
        VOITURES=VOITURES+[[marque_voiture,modele_voiture,titre_voiture,p[0],kill_voiture,carb_voiture,annee_voiture,vues_voit,voiture_url,ville_voit]]
    
    


    num_item=num_item+35
    print(num_item)
    print("URL FINISH ----------------")
```

```python
import pandas as pd
df = pd.DataFrame(VOITURES, columns=["Marque","Modele","Titre annonce","Prix","Kilometrage","Carburant","Année","Nombre de vues","URL","Ville"])
df.to_csv('voitures.csv', index=False)
```
